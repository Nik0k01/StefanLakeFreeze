$$ \rho c \frac{\partial T}{\partial t} +\rho c(\mathbf{u}\cdot\nabla T) = \nabla \cdot (k \nabla T) - \rho L \frac{\partial f_L}{\partial t}  $$
Integrating over the control surface $S^\mathbb{P}$:
$$ \int_{S^\mathbb{P}} \rho c \frac{\partial T}{\partial t} dS + \int_{S^\mathbb{P}} \rho c(\mathbf{u}\cdot\nabla T)dS = \int_{S^\mathbb{P}} \nabla \cdot (k \nabla T) dS - \int_{S^\mathbb{P}}\rho L \frac{\partial f_L}{\partial t} dS $$
We note that:
$$ \mathbf{u}\cdot\nabla T = \nabla \cdot (\mathbf{u}T) - {T(\nabla \cdot \mathbf{u})} $$
The term ${T(\nabla \cdot \mathbf{u})}$ is equal to 0.

Applying the divergence theorem, the surface integral of the convection term become:
$$\int_{S^\mathbb{P}} \rho c(\mathbf{u}\cdot\nabla T)dS = \rho c\oint_{\partial S^\mathbb{P}} (\mathbf{u}T)\cdot \mathbf{n}dl $$
Subsequently, we can decompose the line integral into contributions of 4 terms:
$$ \rho c\oint_{\partial S^\mathbb{P}} (\mathbf{u}T)\cdot \mathbf{n}dl = \rho c\left( \int_{l_{sw}^{se}} (\mathbf{u}T)\cdot \mathbf{n}dl + \int_{l_{se}^{ne}} (\mathbf{u}T)\cdot \mathbf{n}dl + \int_{l_{ne}^{nw}} (\mathbf{u}T)\cdot \mathbf{n}dl + \int_{l_{nw}^{sw}} (\mathbf{u}T)\cdot \mathbf{n}dl \right) $$
Performing the inner product operation: 
$$ \int_{l_{sw}^{se}} (\mathbf{u}T)\cdot \mathbf{n}dl = \frac{\Delta y_{sw}^{se}}{||l_{sw}^{se}||}\int_{l_{sw}^{se}}u_xTdl - \frac{\Delta x_{sw}^{se}}{||l_{sw}^{se}||}\int_{l_{sw}^{se}}u_yTdl $$
The integral could be evaluated by a midpoint rule, however, using a midpoint rule could introduce numerical instabilities. Using an upward scheme, meaning taking the upstream value of the temperature field, leads to a stable solution. Velocity values at the faces are interpolated linearly. 
$$ \frac{\Delta y_{sw}^{se}}{||l_{sw}^{se}||}\int_{l_{sw}^{se}}u_xTdl - \frac{\Delta x_{sw}^{se}}{||l_{sw}^{se}||}\int_{l_{sw}^{se}}u_yTdl = \frac{\Delta y_{sw}^{se}}{||l_{sw}^{se}||} ||l_{sw}^{se}|| u_x\Big |_s T_f - \frac{\Delta x_{sw}^{se}}{||l_{sw}^{se}||} ||l_{sw}^{se}||  u_y\Big |_s T_f  = \Delta y_{sw}^{se}u_x\Big |_s T_f - \Delta x_{sw}^{se} u_y\Big |_s T_f $$
The temperature $T_f$ depends on the flow direction:
$$ T_f = \begin{cases} T_P & \text{if } \Delta y_{sw}^{se}u_x\Big |_s - \Delta x_{sw}^{se} u_y\Big |_s > 0 \\
  T_S & \text{if } \Delta y_{sw}^{se}u_x\Big |_s - \Delta x_{sw}^{se} u_y\Big |_s < 0 \end{cases} $$
It can be written as:
$$ F_S= \rho c(\Delta y_{sw}^{se}u_x\Big |_s - \Delta x_{sw}^{se} u_y\Big |_s) $$
$$a_s = D_S + \max (0, -F_S)$$
$$a_p = D_p + \max (0, F_S) $$

# Velocity field divergance

The velocity field divergence is given as:
$$ \nabla \cdot \mathbf{u} = -\frac{\rho_l-\rho_s}{\rho_l} \frac{\partial f_L}{\partial t} $$
Taking the integral of both sides over the CV:
$$\int_{S^{\mathbb{P}}}\nabla \cdot \mathbf{u}\, dS = \int_{S^{\mathbb{P}}}-\frac{\rho_l-\rho_s}{\rho_l} \frac{\partial f_L}{\partial t}\, dS $$
On the LHS we can apply the divergence theorem, while the RHS is simply the surface of the cell, because all of the terms in the integral can be treated as constant:
$$\oint_{\partial S^{\mathbb{P}}} (\mathbf{u}\cdot \mathbf{n}) dl = -\frac{\rho_l-\rho_s}{\rho_l} \frac{\partial f_L}{\partial t} \cdot S^{\mathbb{P}}$$
LHS will be just fluxes across the borders of the CV:
$$(\Delta y_{sw}^{se}u_x - \Delta x_{sw}^{se} u_y)\Big |_S + (\Delta y_{ne}^{nw}u_x - \Delta x_{ne}^{nw} u_y)\Big |_N + (\Delta y_{se}^{ne}u_x - \Delta x_{se}^{ne} u_y)\Big |_E + (\Delta y_{nw}^{sw}u_x - \Delta x_{nw}^{sw} u_y)\Big |_W = -\frac{\rho_l-\rho_s}{\rho_l} \frac{\partial f_L}{\partial t} \cdot S^{\mathbb{P}}$$

In our simulation there is no flow to the sides, there is only fluid movement due to the movement of the interface, hence:
$$(- \Delta x_{sw}^{se} u_y)\Big |_S + (- \Delta x_{ne}^{nw} u_y)\Big |_N  = -\frac{\rho_l-\rho_s}{\rho_l} \frac{\partial f_L}{\partial t} \cdot S^{\mathbb{P}}$$
